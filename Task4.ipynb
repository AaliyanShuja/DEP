```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler , LabelEncoder
from sklearn.ensemble import IsolationForest
!pip install scikit-learn
from sklearn.metrics import  classification_report, confusion_matrix
```

```python
data=pd.read_csv('/content/CloudWatch_Traffic_Web_Attack.csv')
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
categorical_columns =['src_ip','src_ip_country_code','protocol','dst_ip','rule_names','observation_name','source.name','source.meta','detection_types']
for column in categorical_columns:
  le=LabelEncoder()
  data[column]=le.fit_transform(data[column])
```

```python
data
```

```python
x= data.drop(columns=['creation_time','end_time','time'])
```

```python
x
```

```python
x['src_ip'].plot(kind='hist', bins=40)
plt.title('Plotting src_ip')
plt.xlabel('value')
plt.ylabel('depth')
plt.show()
```

```python
x['src_ip_country_code'].plot(kind='hist', bins=20)
plt.title('Plotting src_ip_country_code')
plt.xlabel('value')
plt.ylabel('depth')
plt.show()
```

```python
scalar = StandardScaler()
x_scaled = scalar.fit_transform(x)
```

```python
x_scaled
```

```python
x_train,x_test=train_test_split(x_scaled,test_size=0.2,random_state=42)
```

```python
model= IsolationForest(contamination=0.1,random_state=42)
model.fit(x_train)
```

```python
y_pred_test =model.predict(x_test)
y_pred_test
```

```python
y_pred_test = [1 if x==1 else 0 for x in y_pred_test]
y_pred_test
```

```python
y_test=[1]*len(y_pred_test)
```

```python
print("Classification_Report")
print(classification_report(y_test,y_pred_test))
```

```python
con_matrix= confusion_matrix(y_test,y_pred_test)
print("Confusion_Matrix: ",con_matrix)
```
