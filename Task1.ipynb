```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('/content/Daily Household Transactions.csv').head()
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df=df.drop_duplicates()
```

```python
df.hist()
```

```python
df.info()
```

```python
colors = df['Income/Expense'].map({'Income': 'blue', 'Expense': 'red'})
```

```python
plt.figure(figsize=(14,6))
plt.scatter(df['Category'], df['Amount'],c=colors, alpha=0.6)
plt.title('Scatter Plot in between Amount and Category')
plt.xlabel('Category')
plt.ylabel('Amount')
plt.show()
```

```python
plt.figure(figsize=(14,6))
plt.scatter(df['Mode'], df['Amount'],c=colors, alpha=0.6)
plt.title('Scatter Plot in between Amount and Mode')
plt.xlabel('Mode')
plt.ylabel('Amount')
plt.show()
```

```python
plt.figure(figsize=(14,6))
plt.scatter(df['Date'], df['Amount'],c=colors, alpha=0.6)
plt.title('Scatter Plot in between Amount over time')
plt.xlabel('Date')
plt.ylabel('Amount')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(df[['Amount']])
df['Cluster'] = kmeans.predict(df[['Amount']])
```

```python
plt.figure(figsize=(14, 7))
plt.scatter(df['Category'], df['Amount'], c=df['Cluster'],alpha=0.5)
plt.title('K-means Clustering of in between Amount and Category')
plt.xlabel('Category')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.colorbar(label='Cluster')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.scatter(df['Mode'], df['Amount'], c=df['Cluster'],alpha=0.5)
plt.title('K-means Clustering of in between Amount and Mode')
plt.xlabel('Mode')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.colorbar(label='Cluster')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.scatter(df['Date'], df['Amount'], c=df['Cluster'],alpha=0.5)
plt.title('K-means Clustering of Amount over Time')
plt.xlabel('Date')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.colorbar(label='Cluster')
plt.show()
```
